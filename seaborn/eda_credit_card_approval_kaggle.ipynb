```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("application_record.csv")
```

```python
df.head()
```

```python
df = df.set_index("ID")
df.head()
```

```python
df.describe()
```

```python
df_employed = df[df["DAYS_EMPLOYED"]<0]
df_employed["DAYS_EMPLOYED"] = -1*df_employed["DAYS_EMPLOYED"]
df_employed["DAYS_BIRTH"] = -1*df_employed["DAYS_BIRTH"]
df_employed.describe()
```

```python
plt.figure(dpi=150)
sns.scatterplot(data=df_employed, x="DAYS_BIRTH", y="DAYS_EMPLOYED")
```

```python
df["AGE_IN_YEARS"] = np.round(df["DAYS_BIRTH"]/365*-1)
df.head()
```

```python
plt.figure(figsize=(16, 6), dpi=100)
sns.histplot(data=df, x="AGE_IN_YEARS", bins=20)
plt.xlabel("Age in Years")
```

```python
df_bottom_half = df.nsmallest(n=int(0.5*len(df)), columns="AMT_INCOME_TOTAL")
plt.figure(figsize=(10, 4), dpi=200)
sns.boxplot(data=df_bottom_half, x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", hue="FLAG_OWN_REALTY")
plt.legend(bbox_to_anchor=(1,1), title="FLAG_OWN_REALTY")
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop(columns=["CODE_GENDER", "FLAG_OWN_CAR", "FLAG_MOBIL", "FLAG_OWN_REALTY", "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", "NAME_FAMILY_STATUS", "NAME_HOUSING_TYPE", "OCCUPATION_TYPE"],axis=1).corr())
```
