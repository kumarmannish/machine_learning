```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.catplot(data=df, x="gender", y="math score", kind="box")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.catplot(data=df, x="gender", y="math score", kind="box", row="test preparation course")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.catplot(data=df, x="gender", y="math score", kind="box", col="test preparation course")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.catplot(data=df, x="gender", y="math score", kind="box", col="test preparation course", row="lunch")
```

```python
grid = sns.PairGrid(data=df, hue="gender")
grid = grid.map_upper(sns.scatterplot)
grid = grid.map_lower(sns.kdeplot)
grid = grid.map_diag(sns.kdeplot)
```
