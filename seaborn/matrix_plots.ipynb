```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("country_table.csv")
```

```python
df.head()
```

```python
df = df.set_index("Countries")
df.head()
```

```python
plt.figure(dpi=150)
sns.heatmap(data=df, linewidths=0.2, annot=True)
```

```python
plt.figure(dpi=150)
sns.heatmap(df.drop("Life expectancy", axis=1), linewidths=0.2, annot=True)
```

```python
plt.figure(dpi=150)
sns.clustermap(df.drop("Life expectancy", axis=1), linewidths=0.2, annot=True)
```
