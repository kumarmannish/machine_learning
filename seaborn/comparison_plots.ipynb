```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score", kind="hex")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score", alpha=0.2)
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score", kind="hist")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score", kind="kde", fill=True)
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.jointplot(data=df, x="math score", y="reading score", kind="kde", fill=True, hue="gender")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.pairplot(data=df)
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.pairplot(data=df, hue="gender")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.pairplot(data=df, hue="gender", corner=True)
```
