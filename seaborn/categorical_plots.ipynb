```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.countplot(data=df, x="division")
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.countplot(data=df, x="division")
plt.ylim(90, 280)
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.countplot(data=df, x="level of education")
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.countplot(data=df, x="level of education", hue="division")
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.barplot(data=df, x="level of education", y="salary", errorbar="sd", estimator=np.mean)
```

```python
plt.figure(figsize=(14, 6), dpi=300)
sns.barplot(data=df, x="level of education", y="salary", errorbar="sd", estimator=np.mean, hue="division")
plt.legend(bbox_to_anchor=(1, 1))
```
